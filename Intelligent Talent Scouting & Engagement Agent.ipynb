```python
import google.generativeai as genai
from google.colab import userdata

# 1. SETUP
try:
    api_key = userdata.get('GOOGLE_API_KEY')
    genai.configure(api_key=api_key)

    # This part AUTOMATICALLY finds the right model name for your key
    available_models = [m.name for m in genai.list_models() if 'generateContent' in m.supported_generation_methods]
    if available_models:
        model_name = available_models[0] # Use the first one that works
        model = genai.GenerativeModel(model_name)
        print(f"✅ System Ready: Using model {model_name}")
    else:
        print("❌ No models found. Please check your API key.")
except Exception as e:
    print(f"❌ Setup Error: {e}")

# 2. THE AGENT LOGIC
def talent_scout_agent(job_description, candidate_data):
    prompt = f"Compare JD: {job_description} with Candidate: {candidate_data}. Provide Match Score, Interest Score, and a short message."
    response = model.generate_content(prompt)
    return response.text

# 3. INTERACTIVE INTERFACE
print("------------------------------------------")
print("🌟 DECCAN AI: CATALYST TALENT SCOUT AGENT 🌟")
print("------------------------------------------")

user_jd = input("STEP 1: Paste JD: ")
user_candidates = input("STEP 2: Paste Candidate: ")

# 4. EXECUTION
try:
    print("\n--- Processing... ---\n")
    print(talent_scout_agent(user_jd, user_candidates))
except Exception as e:
    print(f"Final Error: {e}")
```
